```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
reviews = pd.read_csv('../data/beer_reviews.csv')
```

```python
reviews.head()
```

```python
reviews_dropna = pd.DataFrame(reviews.dropna())
```

```python
reviews_dropna.to_csv('../data/reviews_dropped_nulls.csv')
```

```python
reviews_dropna = reviews_dropna[['review_time', 
                'review_profilename', 
                'beer_style', 
                'beer_abv', 
                'brewery_name', 
                'brewery_id', 
                'beer_name', 
                'beer_beerid', 
                'review_aroma', 
                'review_appearance', 
                'review_palate', 
                'review_taste', 
                'review_overall'
               ]]
```

```python
y = reviews_dropna['review_overall']
X = reviews_dropna.drop(columns='review_overall')

X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.to_csv('../data/X_train.csv')
X_test.to_csv('../data/X_test.csv')
y_train.to_csv('../data/y_train.csv')
y_test.to_csv('../data/y_test.csv')
```
